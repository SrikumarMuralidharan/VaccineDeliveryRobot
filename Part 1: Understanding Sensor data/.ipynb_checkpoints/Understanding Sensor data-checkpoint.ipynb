Goal: Load sensor data with Python and plotting data with Matplotlib. Data from IMU is stored in imudata.txt.

Procedure: <br/>
Step 1 : Import required libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Step 2: Read the text file and store it in imudata variable

```python
imudata = open('imudata.txt','r').read()
```

Step 3: Create a new dictionary variable to store pitch data, count is the key value that needs to be updated for this dictionary.

```python
pitch={}
count=0
```

Step 4: Iterating to filter out the 5th column and store it as pitch data in the dictionary. The key and values for the same are updated here.

```python
for row in imudata.splitlines():
    (key,value)=(count,int(row.split()[4]))  #5th column has pitch angles
    pitch[key]=value  #storing information in this dictionary
    count+=1
#print('pitch',pitch)
```

Step 5: Plotting the raw data (pitch angle in deg) against instances.

```python
plt.plot(pitch.keys(), pitch.values(),label='Raw Data')
plt.xlabel('Instance')
plt.ylabel('Pitch angle (in deg)')
plt.title("Plotting pitch angle of accelerometer ADXL327")
plt.legend(loc="upper right")
plt.show()
```

Step 6: Add the weights for our moving point average methods.

```python
moving_avg_weights = [2,4,8,16,64,128]
```

Step 7: Frame the function to implement weighted averages.

```python
def MovingAverage(data, weights):
    for weight in weights:
        print("Moving weight considered : ", weight)
        MeanList=[]
        TotalMeanList=[]
        StartList=[]
        start_index=0
        Sum_MeanList=0
        Variance = 0
        values = list(data.values())
        keys = list(data.keys())
        
        while start_index<len(values)-weight:
            UpdatedList = values[start_index:start_index+weight]
            UpdatedSum=0;
            for L in UpdatedList:
                UpdatedSum+=L
            UpdatedMean=UpdatedSum/weight
            Sum_MeanList+=UpdatedMean
            MeanList.append(UpdatedMean)
            StartList.append(weight+start_index)
            start_index+=1
        TotalMean=Sum_MeanList/len(MeanList)
        for i in values:
            TotalMeanList.append(TotalMean)
        print("Mean : ", TotalMean)
        for M in MeanList:
            Variance+= (M-TotalMean)**2
        Variance/=len(MeanList)-1
        StandardDeviation=Variance**0.5
        print("Standard deviation : ", StandardDeviation)
        plt.title("Moving average plot vs raw data")
        plt.plot(keys,values,label="Raw Data")
        plt.xlabel('Instance')
        plt.ylabel('Pitch angle (in deg)')
        plt.plot(StartList, MeanList,label="Weighted data")
        plt.plot(keys,TotalMeanList,label="Mean")
        plt.legend()
        plt.show()
```

Step 8: Using function to plot for multiple weights.

```python
MovingAverage(pitch, moving_avg_weights)
```
